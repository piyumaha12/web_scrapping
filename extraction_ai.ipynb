# Extraction Program

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import datetime

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.110 Safari/537.36',
}

EX_URL= 'https://insights.blackcoffer.com/how-do-deep-learning-models-predict-old-and-new-drugs-that-are-successfully-treated-in-healthcare/'
```

```python
# Function for url opening
def open_url(url, headers=headers):
    R = requests.get(url, headers=headers).text
    soup = BeautifulSoup(R, 'lxml')
    return soup


# Title function
def title_extract(soup):
    title = soup.find('h1', class_ = 'entry-title').text
    return title


# Content function
def content_extract(soup):
    contents_whole = soup.find('div', class_ = 'td-post-content')
    # contents = contents_whole.find_all(['p', 'h3', 'li'])
    content_string = ''
    for content in contents_whole:
        content_string+= ' '+content.text
    return content_string
    

def read_excel_file(path, Target_colm,index_col = None):
    df_excel = pd.read_excel(path, index_col= index_col)
    list_of_webs = list(df_excel[Target_colm])
    return list_of_webs


def save_data(list_or_dict, name):
    data_frame_save = pd.DataFrame(list_or_dict)
    time_str = datetime.datetime.now().strftime('%d-%m-%Y_%H-%M-%S')
    name = name + time_str+'.csv'
    data_frame_save.to_csv(name, index = False)
```

```python
PATH = 'H:/Excelr/Project/geetanjali/Input.xlsx'
list_of_webs = read_excel_file(PATH, Target_colm= 'URL', index_col= 'URL_ID')
list_of_webs
```

```python
titles = []
contents = []
for i, web in enumerate(list_of_webs):
    print(f'website {i}')
    soup = open_url(web)
    title = title_extract(soup)
    content = content_extract(soup)
    titles.append(title)
    contents.append(content)
    if (i+1)%50 == 0:
        dict_web = {'title': titles, 'content': contents}
        name = 'with_webs_' + str(i+1)
        save_data(dict_web, name)

dict_web_complete = {'title': titles, 'content': contents}  
save_data(dict_web_complete, 'final_file_')
```
